```python
# 查重複資料
# 获取重复的文档
pipeline = [
    {"$group": {"_id": "$id", "count": {"$sum": 1}}},
    {"$match": {"count": {"$gt": 1}}}
]
duplicate_documents = list(_rapID.collection.aggregate(pipeline))

duplicate_documents
```

```python
# 删除重复的文档，保留第一次出现的文档
for document in duplicate_documents:
    document_id = document['_id']
    first_document = _rapID.collection.find_one({"id": document_id})
    _rapID.collection.delete_many({"id": document_id, "_id": {"$ne": first_document["_id"]}})
```

```python
# 確定有href 來更新 增加 doc的 albums欄位
def add_album_from_response(old_dict, response):
    # init response
    
    # try:
    #     response = backendFunc.requestUrl(artist_album_url, headers, 'json')
    # except:
    #     print('soptify api error')
    #     # update headers
    #     type, token = backendFunc.checkToken('spotify_token', search_ID)
    #     headers = {'Authorization': f"{type} {token}"}
    #     response = backendFunc.requestUrl(artist_album_url, headers, 'json')

    new_dict = old_dict.copy()

    if 'albums' not in new_dict.keys():
        new_dict['total_albums'] = response['total']
        new_dict['albums'] = []
    # else:
    #     new_dict['total_albums'] = new_dict['total_albums']+ response['total']

    for i in range(len(response['items'])): # for each album in item
        one_album = response['items'][i]
        temp_dict = {}
        temp_dict['album_name'] = one_album['name']
        temp_dict['total_tracks_num'] = one_album['total_tracks']
        temp_dict['album_id'] = one_album['id']
        temp_dict['release_date'] = one_album['release_date']
        temp_dict['total_artists'] = one_album['artists']
        temp_dict['album_type'] = one_album['album_type']
        temp_dict['images'] = one_album['images']

        new_dict['albums'].append(temp_dict)     

    return new_dict
```

```python
# 確定有href 來更新 增加 doc的 singles欄位
def add_single_from_response(old_dict, response):
    new_dict = old_dict.copy()

    if 'singles' not in new_dict.keys():
        new_dict['total_singles'] = response['total']
        new_dict['singles'] = []
    
    for i in range(len(response['items'])): # for each album in item
        one_single = response['items'][i]
        temp_dict = {}
        temp_dict['single_name'] = one_single['name']
        temp_dict['total_singles_num'] = one_single['total_tracks']
        temp_dict['single_id'] = one_single['id']
        temp_dict['release_date'] = one_single['release_date']
        temp_dict['total_artists'] = one_single['artists']
        temp_dict['album_type'] = one_single['album_type']
        temp_dict['images'] = one_single['images']

        new_dict['singles'].append(temp_dict)     

    return new_dict
```
